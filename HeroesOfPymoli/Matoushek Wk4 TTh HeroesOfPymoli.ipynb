## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting

```python
#Set up variables
tot_purchases = purchase_data["Price"].count()
uniq_offrd_items = len(purchase_data["Item Name"].unique())
item_avg_price = purchase_data["Price"].mean()
total_revenue = purchase_data["Price"].sum()

#Create Data Frame
gen_prchse_anal = pd.DataFrame({"1) Unique Items Offered":[uniq_offrd_items], 
              "2) Avg Purchase Price":item_avg_price, 
              "3) Number of Purchases":tot_purchases, 
              "4) Total Purchases Revenue":total_revenue})

# Improve formatting before outputting spreadsheet
gen_prchse_anal["1) Unique Items Offered"] = gen_prchse_anal["1) Unique Items Offered"].map("{:,}".format)
gen_prchse_anal["2) Avg Purchase Price"] = gen_prchse_anal["2) Avg Purchase Price"].map("${0:,.2f}".format)
gen_prchse_anal["3) Number of Purchases"] = gen_prchse_anal["3) Number of Purchases"].map("{:,}".format)
gen_prchse_anal["4) Total Purchases Revenue"] = gen_prchse_anal["4) Total Purchases Revenue"].map("${0:,.2f}".format)

gen_prchse_anal
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
hop_dmogrphx = purchase_data.loc[:,['Gender', 'SN', 'Price']]
hop_dmogrphx = hop_dmogrphx.drop_duplicates()
tot_uniq_plyrs = len(hop_dmogrphx['SN'].unique())
sales_by_gender = hop_dmogrphx.groupby('Gender')

#Calculate Variables
sales_count_by_gender = sales_by_gender["Price"].count()
gndr_purchase_totals = sales_by_gender["Price"].sum()
avg_gndr_purchase = sales_by_gender["Price"].mean()

#Create a DataFrame
gender_purchases = pd.DataFrame({"1) Purchases Count by Gender": sales_count_by_gender, 
                                 "2) Total Purchases by Gender": gndr_purchase_totals,
                                 "3) Avg Purchase by Gender": avg_gndr_purchase,
                                 "4) Avg Purchase per Person": avg_gndr_purchase
                                 })

#Format Data Table
gender_purchases["1) Purchases Count by Gender"] = gender_purchases["1) Purchases Count by Gender"].map("{:,}".format)
gender_purchases["2) Total Purchases by Gender"] = gender_purchases["2) Total Purchases by Gender"].map("${0:,.2f}".format)
gender_purchases["3) Avg Purchase by Gender"] = gender_purchases["3) Avg Purchase by Gender"].map("${0:,.2f}".format)
gender_purchases["4) Avg Purchase per Person"] = gender_purchases["4) Avg Purchase per Person"].map("${0:,.2f}".format)

#Display
gender_purchases
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Bin the Age column
purchase_data['hero_age_bins'] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

#Select for unique player id's and group
hop_dmogrphx = purchase_data.loc[:,['SN', 'Age', 'hero_age_bins']]
hop_dmogrphx = hop_dmogrphx.drop_duplicates()
tot_uniq_plyrs = len(hop_dmogrphx['SN'].unique())
hero_age_dmogrfx = hop_dmogrphx.groupby('hero_age_bins')

#Create Variables
plyr_dmogrphx_totals = hero_age_dmogrfx["hero_age_bins"].count()
plyr_dmogrphx_prcnts = plyr_dmogrphx_totals / tot_uniq_plyrs * 100

#Create Data Frame
hero_age_dmogrfx = pd.DataFrame({"1) Player Count by Age Group":  plyr_dmogrphx_totals, 
                                 '2) Player Percentages by Age Group': plyr_dmogrphx_prcnts
                                })
#Format Data Table
hero_age_dmogrfx["2) Player Percentages by Age Group"] = hero_age_dmogrfx["2) Player Percentages by Age Group"].map("%{0:,.2f}".format)

#Display
hero_age_dmogrfx
```
